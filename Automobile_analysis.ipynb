## Data Wrangling

```python
# Evaluating Missing Values
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Counting Missing Values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# Disabling this new warning
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'     
```

```python
# Replacing missing values by mean in column ("normalize-losses,stroke,bore,horsepower,peak-rpm")
avg_normal_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average Normalized loss is",avg_normal_loss)

df['normalized-losses'].replace(np.nan,avg_normal_loss,inplace=True)
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average stroke is",avg_stroke)

df['stroke'].replace(np.nan,avg_stroke,inplace=True)
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0)
print('Average bore is',avg_bore)

df['bore'].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
print('Average peak-rpm is',avg_peak_rpm)

df['peak-rpm'].replace(np.nan,avg_peak_rpm,inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Average horsepower is',avg_horsepower)

df['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
```

```python
# Ridge Regression

# importing PolynomialFeatures from sklearn.preprocessing
from sklearn.preprocessing import PolynomialFeatures

# test data will be used as validation data
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train[['horsepower','curb-weight','engine-size','highway-mpg','symboling','normalized-losses']])
x_test_pr = pr.fit_transform(x_test[['horsepower','curb-weight','engine-size','highway-mpg','symboling','normalized-losses']])
```

```python
# Creating Ridge object setting parameter alpha = 0.1
RidgeModel = Ridge(alpha=1)
```

```python
# fit the model
RidgeModel.fit(x_train_pr, y_train)
```

```python
yhat = RidgeModel.predict(x_test_pr)  # Obtain a prediction
```

```python
# 'tqdm' package is used for creating Progress Bars
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
# plot out value of R^2 for different alphas
width = 12
height = 10
plt.figure(figsize=(width,height))

plt.plot(Alpha,Rsqu_test,label = 'Validation data')
plt.plot(Alpha,Rsqu_train,'r',label='Training data')
plt.xlabel('Alpha')
plt.ylabel('R^2')
plt.legend()
```
